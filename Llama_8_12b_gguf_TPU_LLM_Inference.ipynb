## Step 3: Give the prompt and predict the chat completion

# Model Initalisation

## solved the prompt engineering issue sys prompt defines key objectives carefully  craft it

```python
from llama_cpp import Llama

# Initialize the model
llm = Llama(
    model_path = '/root/.cache/huggingface/hub/models--bartowski--Llama-3.2-3B-Instruct-GGUF/snapshots/5ab33fa94d1d04e903623ae72c95d1696f09f9e8/Llama-3.2-3B-Instruct-f16.gguf',
    n_ctx=60000,
    n_batch=50000,
    n_threads=96,
    n_gpu_layers=0,
    use_mmap=True,
    use_mlock=True,
    offload_kv=True,
    verbose=True
)


# Your input prompt
prompt = """explain thermodynamics .
"""

# System prompt and user template
prompt_template = f'''
SYSTEM:
"Your an helpful ai assistant"

USER: {prompt}

ASSISTANT:
'''
```

```python
# Note U can find balance bw quality ouput wihtin qunatized models some will perform better at q6 than q8 or equal varies according to model size
# Trial and error method find the balance
```

```python
print("2. STREAMED RESPONSE:")
print("=" * 50)

# Generating response using the Llama model with streaming
stream = llm(
    prompt=prompt_template,
    max_tokens=50000,
    temperature=0.8,
    top_p=0.95,
    repeat_penalty=1.2,
    top_k=50,
    echo=False,
    stream=True,
    stop=None
)


response = ''

# Stream the response real-time
print("\nModel response:")
for output in stream:
    if 'choices' in output and len(output['choices']) > 0:
        text_output = output['choices'][0].get('text', '').replace('\r', '')
        response += text_output
        print(text_output, end='', flush=True)

print("\n" + "=" * 50)
```
